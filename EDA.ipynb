```python
import os
import matplotlib.pyplot as plt
import cv2

# Define dataset paths
train_images_folder = "train\images"  # Correct subfolder
test_images_folder = "test\images"

# Get only image files (ignore directories)
def get_image_files(folder):
    return [f for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Get actual image files
train_image_files = get_image_files(train_images_folder)
test_image_files = get_image_files(test_images_folder)

# Print correct counts
print(f"Total Train Images: {len(train_image_files)}")
print(f"Total Test Images: {len(test_image_files)}")

# Debugging: Print first few image filenames
print("Sample Train Images:", train_image_files[:5])
print("Sample Test Images:", test_image_files[:5])

# Function to display images
def show_images(image_folder, image_list, title):
    fig, axes = plt.subplots(1, min(5, len(image_list)), figsize=(15, 5))  # Adjust if < 5 images
    for idx, img_name in enumerate(image_list[:5]):
        img_path = os.path.join(image_folder, img_name)
        if os.path.exists(img_path):
            img = cv2.imread(img_path)
            if img is None:
                print(f"Skipping corrupted image: {img_name}")
                continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            axes[idx].imshow(img)
            axes[idx].axis("off")
        else:
            print(f"Image not found: {img_path}")
    plt.suptitle(title)
    plt.show()

# Show Train Images
show_images(train_images_folder, train_image_files, "Train Images")

# Show Test Images
show_images(test_images_folder, test_image_files, "Test Images")
```
